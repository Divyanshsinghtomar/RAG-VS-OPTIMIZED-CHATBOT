```python
import re
def clean_prices(x):
    if pd.isna(x):
        return None
    else:
        x = re.sub(r"[₹,]", "", str(x))
        return float(x)
```

```python
sales_df["discounted_price"]=sales_df["discounted_price"].apply(clean_prices)
sales_df["actual_price"]=sales_df["actual_price"].apply(clean_prices)
```

```python
def clean_perc(x):
    if pd.isna(x):
        return None
    else:
        x = re.sub(r"[%,]", "", str(x))
        return int(x)
```

```python
sales_df["discount_percentage"]=sales_df["discount_percentage"].apply(clean_perc)
```

```python
import pandas as pd

from src.preprocessing.chunking import chunk_text

df = sales_df

texts = df["about_product"].dropna().tolist()

all_chunks = []

for text in texts:

    chunks = chunk_text(
        text,
        chunk_size=100,
        overlap=20
    )

    all_chunks.extend(chunks)

print(f"Total chunks: {len(all_chunks)}")
```

```python
from src.embeddings.embedder import EmbeddingModel

texts = [
    "great battery life",
    "poor camera quality"
]

embedder = EmbeddingModel()

embeddings = embedder.encode(all_chunks)

print(embeddings.shape) # gives (2, 384) for 384-dimensional embeddings
```

```python
metadata = []

for chunk in all_chunks:

    metadata.append({
        "text": chunk
    })
```

```python
from src.vector_store.faiss_store import FAISSVectorStore

dimension = embeddings.shape[1]

vector_store = FAISSVectorStore(dimension)
```

```python
vector_store.add_embeddings(
    embeddings,
    metadata
)
```

```python
query = "good battery backup phone"

query_embedding = embedder.encode([query])
```

```python
results = vector_store.search(
    query_embedding,
    top_k=5
)
```

```python
for r in results:

    print("Score:", r["score"])
    print(r["metadata"]["text"])
    print("-" * 80)
```
